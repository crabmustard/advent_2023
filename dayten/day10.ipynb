```python
def find_next(grid, y, x, d):

    square = grid[y][x]

    nx = x
    ny = y
    nd = d
    if d == 'u':
        if square == '|':
            nd = 'u'
            ny -= 1
        elif square == '7':
            nd = 'l'
            nx -= 1
        elif square == 'F':
            nd = 'r'
            nx += 1
    elif d == 'd':
        if square == '|':
            ny += 1
            nd = 'd'
        elif square == 'J':
            nd = 'l'
            nx -= 1
        elif square == 'L':
            nd = 'r' 
            nx += 1
    elif d == 'l':
        if square == '-':
            nd = 'l'
            nx -= 1
        elif square == 'F':
            nd = 'd'
            ny += 1
        elif square == 'L':
            nd = 'u'
            ny -= 1
    elif d == 'r':
        if square == '-':
            nd = 'r'
            nx += 1
        elif square == '7':
            nd = 'd'
            ny += 1
        elif square == 'J':
            nd = 'u'
            ny -= 1
    return nd, ny, nx
    
    






test_grid , start = make_grid('day10short.txt')
sy, sx, sxt = start[0], start[1], 'd'
nxt, ny, nx = find_next(test_grid, sy, sx, sxt)

```
